```python
#loading the dataset
video_path = "data/videos/"
video_files_path = []

for file in os.listdir(video_path):
    video_files_path.append(os.path.join(video_path, file))
```

```python
#video files
video_files_path
```

```python
cap = cv2.VideoCapture(video_files_path[0]) #use 0 for live capture via webcam
fps = cap.get(cv2.CAP_PROP_FPS)
print("FPS : ", fps)
count = 0
while True:
    #pull a frame from capture
    ret, frame = cap.read()
    if(ret == True):
        cv2.imshow('Frame', frame)
        cv2.imwrite("frames/frame{}.jpg".format(count), frame)
        count += 1
        #press q on keyboard to exit
        if(cv2.waitKey(25) & 0xFF == ord('q')):
            break
    else:
        break

#when everything is done, release the captured video object    
cap.release()
#close all the frames
cv2.destroyAllWindows()
```

```python
print("Type of capture : ", type(cap))
print("No. of frames captured : ", count)
```

```python
import matplotlib.image as mpimg
frame_path = "frames/"
vid_frames = []
for fr in os.listdir(frame_path):
    vid_frames.append(os.path.join(frame_path, fr))
```

```python
#plotting the fourth frame
img = mpimg.imread(vid_frames[3])
plt.imshow(img)
print("Dimensions of the frame : ", img.shape)
print("Height of the frame : ", img.shape[0])
print("Width of the frame : ", img.shape[1])
print("No. of channels in the frame : ", img.shape[2])
print("Pixel values of the frame : \n", img)
```

```python
#sanity check, all the frames have the same dimension
for i, fr in enumerate(vid_frames):
    frame = mpimg.imread(fr)
    print("Dimension of frame {} : {}".format(i + 1, frame.shape))
```
